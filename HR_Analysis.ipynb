1.WHAT FACTORS CONTRIBUTE MOST TO EMPLOYEE ATTRITION?


```python
sns.histplot(data=df, x='Age', hue='Attrition', multiple='stack', bins=20)
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The analysis reveals that attrition rates are higher among employees in their late 20s to early 40s. This age group might face challenges balancing career growth, family responsibilities, or other personal goals, contributing to a higher turnover rate.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title('Count of Employees by Gender and Attrition')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

The data shows that attrition is higher among male employees compared to female employees. This trend might be influenced by factors such as job roles, workplace expectations, or external opportunities that differ by gender.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='JobRole', hue='Attrition', order=df['JobRole'].value_counts().index)
plt.title('Job Role Distribution by Attrition')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Attrition rates vary significantly across job roles, with the highest attrition observed among Laboratory Technicians, followed by Sales Executives and Research Scientists.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Count of Employees by Department and Attrition')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Attrition is highest in the Research and Development (R&D) department, followed by the Sales department. These findings suggest that employees in these departments may face unique challenges, such as high workload, job stress, or limited advancement opportunities, which could contribute to their decision to leave. Addressing these issues through department-specific initiatives, like training, support programs, and career development opportunities, could help improve retention in these critical areas.

```python


plt.figure(figsize=(12, 6))


sns.histplot(df[df['Attrition'] == 'No']['DailyRate'], kde=True, color='blue', alpha=0.6, label='Not Churned')

# Plot daily rate distribution for employees who did churn
sns.histplot(df[df['Attrition'] == 'Yes']['DailyRate'], kde=True, color='red', alpha=0.6, label='Churned')


plt.title('Daily Rate Distribution by Attrition Status')
plt.xlabel('Daily Rate')
plt.ylabel('Frequency')


plt.legend(title='Attrition Status')

plt.tight_layout()
plt.show()
```

```python
mean_performance = df.groupby('JobSatisfaction')['PerformanceRating'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=mean_performance, x='JobSatisfaction', y='PerformanceRating', palette='viridis')
plt.title('Average Performance Rating by Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Average Performance Rating')
plt.xticks(rotation=45)
plt.show()
```

The analysis shows no significant connection between job satisfaction levels and performance ratings. This suggests that employees with varying levels of job satisfaction tend to have similar performance ratings, indicating that other factors might have a stronger influence on employee performance.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df[df['Attrition'] == 'No']['MonthlyIncome'], kde=True, color='blue', alpha=0.6, label='Not Churned')
sns.histplot(df[df['Attrition'] == 'Yes']['MonthlyIncome'], kde=True, color='red', alpha=0.6, label='Churned')
plt.title('Monthly Income Distribution by Attrition Status')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

The analysis reveals that employees with lower monthly incomes exhibit a higher churn rate. This trend suggests that financial dissatisfaction may lead employees to seek better-paying opportunities, indicating that salary is a significant factor in employee retention. Companies may need to consider competitive compensation strategies to improve employee retention, particularly among lower-income groups.

5. WHAT IMPACT DOES OVERTIME HAVE ON JOB SATISFACTION AND ATTRITION?


```python
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)


df['OverTime_binary'] = df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)


attrition_rate_by_overtime = df.groupby('OverTime_binary')['Attrition'].mean().reset_index()
attrition_rate_by_overtime['OverTime'] = attrition_rate_by_overtime['OverTime_binary'].map({0: 'No', 1: 'Yes'})


plt.figure(figsize=(12, 6))
sns.barplot(data=attrition_rate_by_overtime, x='OverTime', y='Attrition', palette='viridis')
plt.title('Attrition Rate by Overtime')
plt.xlabel('OverTime')
plt.ylabel('Attrition Rate')
plt.xticks(rotation=0)
plt.show()
```

The analysis indicates that attrition rates are significantly higher for employees who work overtime. This suggests that excessive work hours may contribute to employee dissatisfaction and burnout, leading to higher turnover. Organizations should evaluate their overtime policies and consider implementing measures to ensure work-life balance, as this could help retain employees who might otherwise leave due to work-related stress.

```python
df['OverTime_binary'] = df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
avg_job_satisfaction_by_overtime = df.groupby('OverTime')['JobSatisfaction'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=avg_job_satisfaction_by_overtime, x='OverTime', y='JobSatisfaction', palette='viridis')
plt.title('Average Job Satisfaction by Overtime Status')
plt.xlabel('OverTime')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=0)  
plt.show()
```
